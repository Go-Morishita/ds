#### 演習問題2-1の解答

```python
breast_cancer = datasets.load_breast_cancer()
breast_df = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
pd.set_option('display.max_columns', None)
breast_df.describe()
```

#### 演習問題2-2の解答

```python
breast_df[['mean area', 'worst area']].boxplot()
plt.show()
```

```python
excluded = ['mean area', 'worst area']
remaining_cols = [col for col in breast_df.columns if col not in excluded]
breast_df[remaining_cols].boxplot(figsize=(50, 6))
plt.savefig('breast_df_remaining_cols.png')
plt.show()
```

#### 演習問題3の解答

#### 演習問題3-1の解答

```python
breast_df.hist(figsize=(20, 15), bins=20)
plt.show()
```

#### 演習問題3-2の解答

```python
breast_df['class'] = [breast_cancer.target_names[i] for i in breast_cancer.target]
```

```python
breast_df.hist(column='mean radius', by='class', sharex=True, sharey=True, layout=(2,1), bins=20, figsize=(8, 6))
plt.show()
```

#### 演習問題4の解答

```python
import seaborn as sns

selected_cols = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'class']

remaining_cols = [col for col in breast_df.columns if col in selected_cols]
selected_breast_df = breast_df[remaining_cols]
```

```python
sns.pairplot(selected_breast_df, hue='class')
plt.show()
```

```python
selected_cols = ['worst radius', 'worst texture', 'worst perimeter', 'worst area', 'class']
remaining_cols = [col for col in breast_df.columns if col in selected_cols]
selected_breast_df = breast_df[remaining_cols]
```

```python
sns.pairplot(selected_breast_df, hue='class')
plt.show()
```
