```dir()``` そのオブジェクトが持っている属性やメソッドの一覧を確認するための関数

```id()``` オブジェクトの識別子（メモリアドレスのようなもの）

```type()``` その変数のデータ型を調べる関数

## 演習2_2

````python
x = 10
````

````python
type(x)
````

````python
id(x)
````

````python
dir(x)
````

````python
y = x
````

````python
id(x)
````

````python
id(y)
````

````python
y = y - 1
````

````python
print(x)
id(x)
````

````python
print(y)
id(y)
````

変数xとy=xで定義した変数yのidは等しいことを確認した。
その後y=y-1のように変数yに対して引き算を行うと変数yのidに変更があった。
idはメモリのオブジェクトアドレスを参照する関数なので、y=xのように変数が変数を定義した場合は異なる変数でも同じアドレスが割り当てられていて、変数に対して操作が加わるとアドレスが変更される挙動が確認できた。
