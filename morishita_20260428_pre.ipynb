```python
from matplotlib import pyplot as plt
```

```python
def check_prob(m, scale="linear"):
    num = []
    prob = []
    count = 0
    for n in range(1, m + 1):
        result = random.randint(1, 6)
        if result == 1:
            count += 1
        num.append(n)
        prob.append(count / n)
    
    print(f'{m}回サイコロを振ったときに1の目が出る確率: {prob[-1]}')
    plt.plot(num, prob)
    plt.xscale(scale)
    plt.show()
```

```python
check_prob(10)
```

## 事前課題2

### ベルヌーイ分布

2種類のみの結果しか得ることができない試行（ベルヌーイ試行）の結果を0と1で表した分布である。

適用例: コインを1回投げて表が出るか、ある顧客が商品を購入するかしないかなど、結果が「成功か失敗か」「Yes か No か」の2択になる場面に使える。

### 二項分布

成功確率pのベルヌーイ試行をn回独立に繰り返したときに、成功する回数の分布である。

適用例: 「100人にDMを送って何人が反応するか」「10回打席に立ってヒットが何本出るか」のように、同じ試行を複数回繰り返したときの成功回数を考えたい場面に使える。

### ポアソン分布

単位時間あたりに平均λ回発生する事象について、実際に発生する回数の分布である。

適用例: 「1時間に何人のお客がコンビニに来るか」「1日に何件の問い合わせメールが届くか」のように、めったに起きないがランダムに発生するイベントの件数を予測したい場面に使える。

### 多項分布

二項分布をk種類の結果に拡張した分布であり、各カテゴリの発生確率がそれぞれ与えられているとき、n回の独立試行のうち各カテゴリが何回発生するかの分布である。

適用例: 「サイコロを100回振ったときに1〜6の目がそれぞれ何回出るか」「選挙でA〜D候補にそれぞれ何票入るか」のように、複数の選択肢それぞれの回数を同時に扱いたい場面に使える。

### 指数分布

単位時間あたりに平均λ回発生する事象について、次に事象が発生するまでの待ち時間の分布である。

適用例: 「次のお客が来店するまでの待ち時間」「電球が切れるまでの寿命」のように、ある事象が発生するまでにかかる時間や間隔をモデル化したい場面に使える。
